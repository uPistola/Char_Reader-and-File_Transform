```python
#Lucas Braga Schuck, João Gabriel Pitol, Rafael Vargas, Henrique Zan Grande
import math
#Função usando a série de Taylor
def ArcoSeno(x):
    if x < -1 or x > 1:
        raise ValueError("O valor de x deve estar entre -1 e 1.")  
    arco = 0
    for n in range(0, 1000):
        numerador = math.factorial(2*n)
        denominador = (4**n) * (math.factorial(n)**2) * (2*n + 1)
        termo = (numerador / denominador) * (x**(2*n + 1))
        arco += termo
        angulo = arco*180/math.pi
    return angulo
```

```python
def DescobrirAngulo(distancia,altura):
  hip = math.sqrt((pow(distancia,2)+ pow(altura,2)))
  seno=altura/hip
  angulo = ArcoSeno(seno)
  return angulo
```

```python
distanciaFoguete = int(input("Insira a distancia do foguete:" ))
alturaFoguete = int(input("Insira a altura do foguete:" ))
angulo = DescobrirAngulo(distanciaFoguete,alturaFoguete)
print(round(angulo,2))
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Definindo um triangulo 3,4,5
x1=np.array([0,distanciaFoguete,distanciaFoguete,0]) #array de pontos
y1=np.array([0,0, alturaFoguete,0]) #(0,0),(4,0),(4,3),(0,0)
# desenhando a figura e os eixos
fig1=plt.figure(1)
# Triângulo 3,4,5
plt.plot(x1,y1,color=[0/255,100/255,200/255]) #plota os segmentos de reta
# deteminando o limite do eixos na figura

# desenhando uma grade cinza para servir de base
plt.grid(axis='both',which='major',color=(0.8, 0.8, 0.8, 0.5), linestyle='-', linewidth=1)
plt.minorticks_on()
plt.grid(axis='both',which='minor',color=(0.8, 0.8, 0.8, 0.5), linestyle=':', linewidth=1)

plt.text(2,0,round(angulo,2),fontsize=15) # ângulo menor
plt.text((distanciaFoguete-5),3,'90º',fontsize=15)
plt.text(48,(alturaFoguete-20),(round(90-angulo,2)),fontsize=15) #ângulo maior
plt.text((distanciaFoguete/2),3,'Distancia',fontsize=15)
plt.text((distanciaFoguete-10),(alturaFoguete/2),'Altura ',fontsize=15)
plt.text(7,(alturaFoguete/2),'Hipotenusa (H)',fontsize=15)
plt.show() # mostra figura na tela
```
