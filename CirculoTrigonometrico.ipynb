```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
pi = 3.14
graus = float(input("Insira o angulo em graus: "))
theta = np.radians(graus)

seno = np.sin(theta)
cosseno = np.cos(theta)
tangente = np.tan(theta)

x2 = np.cos(theta + pi)
y2 = np.sin(theta + pi)
```

```python
fig, ax = plt.subplots()

#Desenha um circulo com centro em 0,0
circle = plt.Circle((0,0), 1, fill=False)
ax.add_artist(circle)

#Desenha os eixos X e Y
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')

ax.axhline(y=seno, color='r')
ax.axvline(x=cosseno, color='r')

#Plota um Ponto no circulo correspondente ao circulo
ax.plot(cosseno, seno, 'ro')

linhaTangente = plt.Line2D([cosseno,x2], [seno,y2], color='r')

#Cria a Definição dos eixos X e Y
ax.set_xlabel('x')
ax.set_ylabel('y')

#Define os Limites dos eixos
ax.set_xlim(-1.1, 1.1)
ax.set_ylim(-1.1, 1.1)
ax.add_artist(linhaTangente)

#Mostra o Plot
plt.show()
```
